**1. State the Hypotheses statement:**



the null (H0) and alternative (H1) hypotheses for the weekly operating cost. The null hypothesis will state that the mean weekly cost is equal to the theoretical model, and the alternative hypothesis will state that the mean weekly cost is higher than the theoretical model.


**2. Calculate the Test Statistic:**

Use the following formula to calculate the test statistic (t):
where:
•	ˉxˉ = sample mean weekly cost (Rs. 3,050)
•	μ = theoretical mean weekly cost according to the cost model (W = $1,000 + $5X for X = 600 units)
•	σ = 5*25 units

```python
x=600

theoretical_weekly = 1000+5*x

std_dev=25*5

print(f"weekly operating cost: ${theoretical_weekly}")
print(f'standard deviation :${std_dev}')

sample_mean_cost=3050

sample_size=25

z_score=(sample_mean_cost-theoretical_weekly)/sample_size
print(f'z score :{z_score}')
```

**3. Determine the Critical Value:**

Using the alpha level of 5% (α = 0.05), determine the critical value from the standard normal (Z) distribution table.


```python
import scipy.stats as stats

alpha=0.05
critical_z_value=stats.norm.ppf(1-alpha)
print(f'critical z value :{critical_z_value}')
```


**4. Make a Decision:**

Compare the test statistic with the critical value to decide whether to reject the null hypothesis.


```python
if z_score > critical_z_value:
    verdict = 'reject Null(H0) hypothesis'
    conclusion = 'There is strong evidence to support the claim that weekly operating costs are higher than the model suggests.'
else:
    verdict = 'fail to reject Null(H0) hypothesis'
    conclusion = 'There is no strong evidence to support the claim that weekly operating costs are higher than the model suggests.'
print(verdict)
print(conclusion)
```
