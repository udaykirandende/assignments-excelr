**Data Preprocessing:**

1.	Impute missing values.

2.	Encode categorical variables using one-hot encoding or label encoding.

```python
age_median=df['Age'].median()
df['Age'].fillna(age_median,inplace=True)
fare_median=df['Fare'].median()
df['Fare'].fillna(fare_median,inplace=True)
embarked_mode=df['Embarked'].mode()[0]
df['Embarked'].fillna(embarked_mode,inplace=True)
```

```python
df['Age'].isnull().sum()
df['Fare'].isnull().sum()
df['Embarked'].isnull().sum()
```

converting categoricalcolumns into numerical

```python
df=pd.get_dummies(df,columns=['Sex','Embarked','Pclass'],drop_first=True)
df
```

```python
df.shape
```

standardizing the data

```python
import re
df['Title']=df['Name'].apply(lambda x:re.search(r'([A-Za-z]+)\.',x).group(1))
df['Title'] = df['Title'].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona', 'Mlle', 'Ms', 'Mme'],
                                   ['Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Rare', 'Miss', 'Miss', 'Mrs'])
df['Title'].value_counts()
```

```python
df['Deck']=df['Cabin'].str[0].fillna("Unknown")
df['Deck'].value_counts()
```

```python
df['FamilySize']=df['SibSp']+df['Parch']+1
df['FamilySize'].value_counts()
```

```python
df.info()
```

```python
df.drop(['Name','Ticket','Cabin','SibSp','Parch'],axis=1,inplace=True)
print(df.columns)
```

**Building Predictive Models:**

1.	Split the preprocessed dataset into training and testing sets.
2.	Choose appropriate evaluation metrics (e.g., accuracy, precision, recall, F1-score) for model evaluation.

```python
x=df.drop(['Survived'],axis=1)
y=df['Survived']
```

```python
x=x.drop('PassengerId',axis=1)
```

```python
print("x_shape:",x.shape)
print("y_shape:",y.shape)
```

```python
train_indices=y[y.notnull()].index
test_indices=y[y.isnull()].index
```

```python



x_train=x.loc[train_indices]
y_train=y.loc[train_indices]
x_test=x.loc[test_indices]

print("x_train_shape:",x_train.shape)
print("y_train_shape:",y_train.shape)
print("x_test_shape:",x_test.shape)
```

```python
!pip install xgboost
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score,roc_auc_score
```

```python
lgbm=LGBMClassifier(random_state=42)
param_grid={
    'n_estimators':[100,200],
    'max_depth':[5,7],
    'learning_rate':[0.05,0.1],
    'num_leaves':[20,31]
}
gs=GridSearchCV(estimator=lgbm,param_grid=param_grid,cv=3,scoring='accuracy',n_jobs=-1,verbose=1)
gs.fit(x_train,y_train)

print('best hyperparameter:',gs.best_params_)
print('best score:',gs.best_score_)

best_lgbm=gs.best_estimator_

y_train_pred=best_lgbm.predict(x_train).astype(int)
y_test_pred=best_lgbm.predict(x_test).astype(int)

print('accuracy_score:',accuracy_score(y_train,y_train_pred))
print('f1_score:',f1_score(y_train,y_train_pred))
print('precision_score:',precision_score(y_train,y_train_pred))
print('recall_score:',recall_score(y_train,y_train_pred))
print("roc_auc:",roc_auc_score(y_train,y_train_pred))
```

```python
xgb=XGBClassifier(random_state=42,eval_metric="logloss")

param_grid={
    'n_estimators':[100,200],
    'max_depth':[3,5],
    'learning_rate':[0.05,0.1],
    'subsample':[0.8,1.0]
}

grid_search_xgb=GridSearchCV(estimator=xgb,param_grid=param_grid,cv=3,scoring='accuracy',n_jobs=-1,verbose=1)
grid_search_xgb.fit(x_train,y_train)

print('best hyperparameter:',grid_search_xgb.best_params_)
print('best score:',grid_search_xgb.best_score_)

best_xgb=grid_search_xgb.best_estimator_

y_train_pred_xgb=best_xgb.predict(x_train).astype(int)
print("xgboost evalutions")
print('accuracy_score:',accuracy_score(y_train,y_train_pred_xgb))
print('f1_score:',f1_score(y_train,y_train_pred_xgb))
print('precision_score:',precision_score(y_train,y_train_pred_xgb))
print('recall_score:',recall_score(y_train,y_train_pred_xgb))
print("roc_auc:",roc_auc_score(y_train,y_train_pred_xgb))


```

**Comparative Analysis:**

1.	Compare the performance metrics (e.g., accuracy, precision, recall) of LightGBM and XGBoost models.

**COMPARISION OF LGBM AND XGB**

```python
lgbm_metrics = {
    'Accuracy': accuracy_score(y_train, y_train_pred),
    'Precision': precision_score(y_train, y_train_pred),
    'Recall': recall_score(y_train, y_train_pred),
    'F1-Score': f1_score(y_train, y_train_pred),
    'ROC-AUC': roc_auc_score(y_train, y_train_pred)
}

xgb_metrics = {
    'Accuracy': accuracy_score(y_train, y_train_pred_xgb),
    'Precision': precision_score(y_train, y_train_pred_xgb),
    'Recall': recall_score(y_train, y_train_pred_xgb),
    'F1-Score': f1_score(y_train, y_train_pred_xgb),
    'ROC-AUC': roc_auc_score(y_train, y_train_pred_xgb)
}

metrics_df = pd.DataFrame({
    'LightGBM': lgbm_metrics,
    'XGBoost': xgb_metrics
})
```

```python
display(metrics_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

metrics_df_transposed = metrics_df.transpose()
fig, ax = plt.subplots(figsize=(12, 6))
metrics_df_transposed.plot(kind='bar', ax=ax)
ax.set_title('Model Performance Comparison (Training Data)')
ax.set_ylabel('Score')
ax.set_xticks(range(len(metrics_df_transposed.index)))
ax.set_xticklabels(metrics_df_transposed.index, rotation=45)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
