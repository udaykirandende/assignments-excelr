```python
import pandas as pd
```

```python
data={

' Satisfaction':['Very Satisfied','Satisfied','Neutral','Unsatisfied','Very Unsatisfied','total'],
'smart Thermostat':[50,80,60,30,20,240],
'smart Light':[70,100,90,50,50,360],
'total':[120,180,110,80,70,600]


}
df=pd.DataFrame(data)
df
```

```python
contingency_table=df[['smart Thermostat','smart Light']].iloc[:-1]
contingency_table
```

2. Compute the Chi-Square Statistic:

3. Determine the Critical Value:
Using the significance level (alpha) of 0.05 and the degrees of freedom (which is the number of categories minus 1)


```python
import scipy.stats as stats
```

calculations

```python
n=len(contingency_table)
degree_of_freedom= n-1

alpha=0.05
chi2_stat,p_value,df_chi2,expected_freq=stats.chi2_contingency(contingency_table)
critical_value=stats.chi2.ppf(1-alpha,df_chi2)
print('Critical Value (alpha=0.05,df=4):',critical_value)
print('Chi-Square Statistic:',chi2_stat)
print('P-value:',p_value)
print('Degrees of Freedom:',df_chi2)
print('Expected Frequencies:\n',expected_freq)
print('Significance Level (alpha):',alpha)
print('n_value:',n)
```

**4. Make a Decision:**

Compare the Chi-Square statistic with the critical value to decide whether to reject the null hypothesis.


```python
#'comparision between chi2_stat and critical value'

if chi2_stat > critical_value:
    verdict = "Reject the null hypothesis (H0)"
    conclusion = "There is a statistically significant relationship between customer satisfaction levels and smart device types."
else:
    verdict_chi2 = "Fail to reject the null hypothesis (H0)"
    conclusion_chi2 = "There is no statistically significant relationship between customer satisfaction levels and smart device types."

#'comparision b/w p_value and alpha'

if p_value < alpha:
    verdict_pvalue = "Reject the null hypothesis (H0)"
    conclusion_pvalue = "There is a statistically significant relationship between customer satisfaction levels and smart device types."
else:
    verdict_pvalue = "Fail to reject the null hypothesis (H0)"
    conclusion_pvalue = "There is no statistically significant relationship between customer satisfaction levels and smart device types."
print(f'verdict_chi2      :{verdict}')
print(f'\nconclusion        :{conclusion}')
print(f'\nverdict_pvalue    :{verdict_pvalue}')
print(f'\nconclusion_pvalue :{conclusion_pvalue}')
```

```python
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"Critical Value (alpha={alpha}, df={df_chi2}): {critical_value:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"\nConclusion based on Chi-Square statistic: {conclusion_chi2}")
print(f"Conclusion based on P-value: {conclusion_pvalue}")
```
