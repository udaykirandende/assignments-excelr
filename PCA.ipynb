# Task 2: Dimensionality Reduction with PCA:

**1.	Standardize the features to ensure they have a mean of 0 and a standard deviation of Implement PCA to reduce the dimensionality of the dataset.**

**2.	Determine the optimal number of principal components using techniques like scree plot or cumulative explained variance.**

**3.	Transform the original dataset into the principal components.**


```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
df_scaled=ss.fit_transform(df)
```

```python
df_scaled
```

```python
from sklearn.decomposition import PCA
pca = PCA()
principal_components = pca.fit_transform(df_scaled)
print(principal_components.shape)
```

```python
explained_variance_ratio = pca.explained_variance_ratio_
print('explained variance ratio per pca:',explained_variance_ratio)


cumulative_explained_variance = np.cumsum(pca.explained_variance_ratio_)
print('cumulative_explained_variance:',cumulative_explained_variance)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance Plot')
plt.axhline(y=0.95, color='r', linestyle='--', label='95% Explained Variance')
plt.legend(loc='best')
plt.grid(True)

plt.tight_layout()
plt.show()
```

cumulative explained variance plot

Approximately 3 components explain about 67% of the variance.

Approximately 6 components explain about 85% of the variance.

Approximately 9-10 components explain about 95% of the variance.

```python
principal_components_df = pd.DataFrame(data=principal_components, columns=[f'PC_{i+1}' for i in range(principal_components.shape[1])])
print(principal_components_df.shape)
display(principal_components_df.head())
```

# Task 3: Clustering with Original Data:

1.	Apply a clustering algorithm (e.g., K-means) to the original dataset.

2.	Visualize the clustering results using appropriate plots.

3.	Evaluate the clustering performance using metrics such as silhouette score or Davies–Bouldin index.


```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score
```

```python
kmeans = KMeans(n_clusters=3, random_state=42,n_init=10)
kmeans.fit(df)
```

```python
df['Cluster'] = kmeans.labels_
print("Cluster labels assigned to DataFrame.")
display(df.head())
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x=principal_components[:, 0], y=principal_components[:, 1], hue=df['Cluster'], palette='viridis', s=100, alpha=0.8)
plt.title('K-means Clustering on Original Data (PCA Components)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
silhouette = silhouette_score(df, kmeans.labels_)
davies_bouldin = davies_bouldin_score(df, kmeans.labels_)

print(f"Silhouette Score: {silhouette:.3f}")
print(f"Davies-Bouldin Index: {davies_bouldin:.3f}")
```

# Task 4: Clustering with PCA Data:

1.	Apply the same clustering algorithm to the PCA-transformed dataset.

2.	Visualize the clustering results obtained from PCA-transformed data.

3.	Compare the clustering results from PCA-transformed data with those from the original dataset.


```python
kmeans_pca=KMeans(n_clusters=3,random_state=42,n_init=10)
kmeans_pca.fit(principal_components)
df['pca_clusters']=kmeans_pca.labels_
print('pca transformed data')
display(df.head())
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x=principal_components[:, 0], y=principal_components[:,1], hue=df['pca_clusters'], palette='viridis', s=100, alpha=0.8)
plt.title('K-means Clustering on PCA Transformed Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
silhouette_pca = silhouette_score(principal_components, df['pca_clusters'])
davies_bouldin_pca = davies_bouldin_score(principal_components, df['pca_clusters'])

print(f"Silhouette Score (PCA): {silhouette_pca:.3f}")
print(f"Davies-Bouldin Index (PCA): {davies_bouldin_pca:.3f}")
```

### Practical Implications of PCA and Clustering
*   **PCA**: Useful for reducing computational complexity, denoising data, and visualizing high-dimensional data by identifying underlying principal components.
*   **Clustering**: Effective for uncovering hidden structures or segments within data, which can lead to targeted strategies or better understanding of data characteristics.
*   **Combined Use**: PCA can precede clustering to improve efficiency or mitigate the curse of dimensionality, but its impact on clustering performance is dataset-dependent.

### Recommendations for When to Use Each Technique
*   **Use Original Data for Clustering**: When the inherent structure in the full feature space provides better cluster separation and interpretability, as observed in this analysis.
*   **Use PCA for Dimensionality Reduction (prior to clustering) when**:
    1.  Dealing with very high-dimensional data where computational cost is a major concern.
    2.  There is significant multicollinearity among features.
    3.  Visualizing data and needing fewer dimensions to represent most of the variance.
    4.  The reduced dimensionality enhances (or at least does not significantly deter) the clustering performance, which should be evaluated empirically.
